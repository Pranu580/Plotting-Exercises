```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

## **Problem 1**

```python
# Drawing 300 random points from a normal distribution
data = np.random.randn(300)  # Standard normal distribution (mean=0, std=1)

# Creating the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)

# Calculating the mean and median
mean = np.mean(data)
median = np.median(data)

# Adding annotations for the mean and median
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')

# Adding labels and title with appropriate font sizes
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram of 300 Random Points (Normal Distribution)', fontsize=16)

# Adding a legend and grid
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# Showing the plot
plt.tight_layout()
plt.show()
```

## **Problem 2**

```python
# Loading data from the provided Parquet file URL
url = "https://star.herts.ac.uk/~kuhn/DHV/exercises_problem2.parquet"

# Reading the Parquet file into a DataFrame
data = pd.read_parquet(url)

# Displaying the first few rows to understand its structure
data.head()
```

```python
# Setting seaborn style for better plot aesthetics
sns.set(style="whitegrid")

# Histograms for var1, var2, and var3
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(data['var1'], kde=True, color='skyblue', bins=20)
plt.title('Distribution of var1')

plt.subplot(1, 3, 2)
sns.histplot(data['var2'], kde=True, color='salmon', bins=20)
plt.title('Distribution of var2')

plt.subplot(1, 3, 3)
sns.histplot(data['var3'], kde=True, color='lightgreen', bins=20)
plt.title('Distribution of var3')

plt.tight_layout()
plt.show()
```

```python
# Boxplots for var1, var2, and var3 grouped by class
sns.set(style="whitegrid")

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(x='class', y='var1', data=data, palette='Set2')
plt.title('Boxplot of var1 by class')

plt.subplot(1, 3, 2)
sns.boxplot(x='class', y='var2', data=data, palette='Set2')
plt.title('Boxplot of var2 by class')

plt.subplot(1, 3, 3)
sns.boxplot(x='class', y='var3', data=data, palette='Set2')
plt.title('Boxplot of var3 by class')

plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
correlation_matrix = data[['var1', 'var2', 'var3']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Class Distribution Plot
plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=data, palette='Set2')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.kdeplot(data=data, x='var1', hue='class', fill=True, palette='Set2')
plt.title('Density Plot of var1 by class')

plt.subplot(1, 3, 2)
sns.kdeplot(data=data, x='var2', hue='class', fill=True, palette='Set2')
plt.title('Density Plot of var2 by class')

plt.subplot(1, 3, 3)
sns.kdeplot(data=data, x='var3', hue='class', fill=True, palette='Set2')
plt.title('Density Plot of var3 by class')

plt.tight_layout()
plt.show()
```
